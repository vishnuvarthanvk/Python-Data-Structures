### 1. Retrieve data from web page using URLLIB and print the frequency of words from that page.

```python
import urllib.request
counts=dict()
web=urllib.request.urlopen('https://en.wikipedia.org/wiki/Lok_Sabha')
for line in web:
    words=line.decode().split()
    for word in words:
        counts[word]=counts.get(word,0)+1
for i in counts.items():
    print(i)
```

### 2. Retrieve and display all hyperlinks (ie., HREF attribute) from a webpage using BeautifulSoup.

```python
import urllib.request,urllib.parse,urllib.error
from bs4 import BeautifulSoup
import ssl
ctx=ssl.create_default_context()
ctx.check_hostname=False
ctx.verify_mode=ssl.CERT_NONE
url=('https://en.wikipedia.org/wiki/Lok_Sabha')
html=urllib.request.urlopen(url,context=ctx).read()
soup=BeautifulSoup(html,'html.parser')
tags=soup('a')
for tag in tags:
    print(tag.get('href',None))
```

### 5. Crawl Weather of a City and Display

```python
import requests
from bs4 import BeautifulSoup
page = requests.get('http://www.weather.com')
page.content
bs4 = BeautifulSoup(page.content,'html.parser')
bs4.find_all('p')
(bs4.find_all(class_='DetailSummary--DetailsSummary--QpFD'))
days=[bs4.find_all('h2')[day].get_text() for day in range(len(bs4.find_all('h2')))]
tempr=[bs4.find_all(class_="DetailsSummary--temperature--3FMlw")[temp].get_text() for temp in range(len(bs4.find_all(class_="DetailsSummary--temperature--3FMlw")))]
days = days[1:14]
tempr = tempr[1:14]
from IPython.display import display
import pandas as pd
b = {"Days":days,"Temperature":tempr}
weather = pd.DataFrame.from_dict(b, orient = 'index')
weather = weather.transpose()
weather
```
