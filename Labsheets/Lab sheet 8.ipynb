### 1. Search for lines that contain 'From' and print them

```python
import re
f=open("D:\\mbox-short.txt",'r+')
for line in f:
    line=line.strip()
    if re.search("from",line):
        print(line)
f.close()
```

### 2.Search for lines that starts with 'From' and print them

```python
f1=open("D:\\mbox-short.txt",'r+')
for i in f1:
    i=i.rstrip()
    if re.search('^From',i):
        print(i)
f1.close()
```

### 3.Search for lines that start with 'F', followed by2 characters, followed by 'm:'

```python
import re
f2 = open('D:\\mbox-short.txt')
for line1 in f2:
    line1 = line1.rstrip()
    if re.search('^F..m:', line1):
        print(line1)
```

### 4.Search for lines that start with From and have an at sign and print them

```python
f3 = open('D:\\mbox-short.txt')
for line in f3:
    line = line.rstrip()
    if re.search('^From:.+@', line):
        print(line)
```

### 5. Search for lines that have an at sign between characters and print using findall()

```python
import re
f4 = open('D:\\mbox-short.txt')
for line2 in f4:
    line2 = line2.rstrip()
    a = re.findall('\S+@\S+', line2)
    if len(a)>0:
        print(a)
        
```

### 6.Search for lines that have an at sign between characters .The characters must be a letter or number and print them

```python
import re
f5 = open('D:\\mbox-short.txt')
for line in f5:
    line = line.rstrip()
    x = re.findall('[a-zA-Z0-9]\S*@\S*[a-zA-Z]', line)
    if len(x) > 0:
        print(x)
```

### 7.Search for lines that start with 'X' followed by any non  whitespace characters and ':' followed by a space and any number. The number can include a decimal.

```python
import re
f6 = open('D:\\mbox-short.txt')
for line in f6:
    line = line.rstrip()
    if re.search('^X\S*: [0-9.]+', line):
        print(line)
```

### 8.Search for lines that start with 'X' followed by any non whitespace characters and ':' followed by a space and any number. The number can include a decimal. Then print the number if it is greater than zero.

```python
f7 = open('D:\\mbox-short.txt')
for line in f7:
    line = line.rstrip()
    x = re.findall('^X\S*: ([0-9.]+)', line)
    if len(x) > 0:
        print(x)
```

### 9.Search for lines that start with 'Details: rev=' followed by numbers Then print the number if one is found

```python
f8 = open('D:\\mbox-short.txt')
for line in f8:
    line = line.rstrip()
    x = re.findall('^Details:.*rev=([0-9]+)', line)
    if len(x) > 0:
        print(x)
```

### 10.Search for lines that start with From and a character followed by a two digit number between 00 and 99 followed by ':'Then print the number if one is found

```python
f9 = open('D:\\mbox-short.txt')
for line in f9:
    line = line.rstrip()
    x = re.findall('^From .* ([0-9][0-9]):', line)
    if len(x) > 0:
        print(x)
```

```python
import re
import sys

def extract_names(filename):
    names=[]
    
    f=open(filename,'r')
    txt=f.read()
    year_match=re.search(r'Popularity\sin\s(\d\d\d\d)',txt)
    if not year_match:
        sys.stderr.write('Couldn\'\'t find the year!\n\n')
        sys.exit(1)
    year=year_match.group(1)
    names.append(year)
    tuples=re.findall(r'<td>(\d+)</td><td>(\w+)</td>\<td>(\w+)</td>',txt)
    names_to_rank={}
    for rank_tuple in tuples:
        (rank,boyname,girlname)=rank_tuple
        if boyname not in names_to_rank:
            names_to_rank[boyname]=rank
        if girlname not in names_to_rank:
            names_to_rank[girlname]=rank
    sorted_name=sorted(names_to_rank.keys())
    for name in sorted_name:
        names.append(name+" "+names_to_rank[name])
    return (names) 
    
extract_names("D:\\baby1990.html")
```
