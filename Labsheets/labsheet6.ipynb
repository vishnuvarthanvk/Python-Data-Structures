```python
#2
f=open("marks.txt",'a+')
n=int(input("How many marks?:"))
li=[]
for i in range(n):
    s=float(input("Enter the mark:")) 
    li.append(s)
f.write(str(li))
f.close()
f1=open("marks.txt",'r')
print(f1.read())
print("Highest Mark is",max(li))
li.sort(reverse=True)
print("Top three marks are",li[:3])
print("Lowest Three marks are,",li[-3:])
f1.close()
```

```python
#3
while True:
    f=open("stock_price.txt","a")
    sto=input("Enter the stock name:")
    f.write(sto)
    f.write('\t')
    for i in range(5):
        p=input("Price:")
        f.write(p)
        f.write('\t')
    f.write('\n')
    q=input("Do you want to continue('D' for done)")
    if q=='D':
        break
f.close()  
```

```python
for sti in open("stock_price.txt",'r').readlines():
    lis=[]
    avg=0
    c=sti.split()
    for j in range(1,5):
        lis.append(int(c[j]))
        maxi=max(lis)
        ind_maxi=lis.index(maxi)
        mini=min(lis)
        ind_min=lis.index(mini)
        avg=sum(lis)/5
    print("Maximum Price",maxi,"at day",ind_maxi+1)
    print("Minimum Price",mini,"at day",ind_min+1)
    print("Average price",avg)
```

```python
print("2.Count the number of lines in a text file:")
file = open('alice.txt',"r")
Counter = 0
a = file.read()
a.split("\n")
for i in a:
    if i:
        Counter += 1
print("Number of lines in the text file:",Counter)
print("\n")
```

```python
print("3.Count the number of unique words in a file:")
num_words = 0
c = open("alice.txt", 'r')
for line in c:
    words = line.split()
    num_words += len(words)
print("Number of words:",num_words)
c.close()
print("\n")
```

```python
print("4.Find Find frequency of words in a given file:")
fname = input('Enter the file name: ')
try:
    fhand = open(fname)
    counts = dict()
    for line in fhand:
        words = line.split()
        for word in words:
            if word in counts:
                counts[word] += 1
            else:
                counts[word] = 1
    print(counts)
except:
    print('File cannot be opened:', fname)
print("\n")
```
