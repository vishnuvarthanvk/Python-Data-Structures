# Web Scraping


## To extract the data from webpage and to find the types of local bodies

```python
import requests
from bs4 import BeautifulSoup
Link = "https://en.wikipedia.org/wiki/List_of_towns_in_India_by_population"
Link_text = requests.get(Link).text
soup = BeautifulSoup(Link_text, 'lxml')

```

```python
#To print the title of the web page without any tags
print(soup.title.string)
```

```python
#To print the required html tags from webpage
our_table = soup.find('table', class_= 'wikitable sortable')
print(our_table)
```

```python
#To scrape particular tag inside a tag
our_table = soup.find('table')
table_links = our_table.find_all('a')
print(table_links)
```

```python
#Extracting the data
types = []
for links in table_links:
    types.append(links.get('title'))
import pandas as pd
df = pd.DataFrame(types)
print(df)
```
